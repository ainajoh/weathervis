```python
from imetkit.check_data import check_data
from imetkit.get_data import get_data
import numpy as np
from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
campaigndates = []
camp = datetime(year=2020, month=2, day=17, hour=0)
camp_end = datetime(year=2020, month=3, day=17, hour=0)

while camp < camp_end:
    campaigndates.append(camp)
    camp +=timedelta(hours=12)
    
```

```python
for date in campaigndates[0:1]:
    date = date.strftime("%Y%m%d%H")
    model ="AromeArctic"
    param = ["x_wind_10m"]
    print(date)
    check = check_data(model=model, date=date, param=param)
    data = get_data(file=check.file, model=model, date=date, param=param, step=[0])
    data.retrieve()
    make_plot(data)
```

```python
def make_plot(data):
    crs = ccrs.PlateCarree( )
    fig = plt.figure()
    ax = plt.axes(projection=crs)
    ax.coastlines()
    CF = ax.contourf(data.longitude, data.latitude, data.x_wind_10m[0,0,:,:], transform = crs ) #filled contours
    #plt.show()
```
